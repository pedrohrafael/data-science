```python
count_status = olist_orders_dataset.groupby('order_status') \
                                   .agg({'order_id':'count'}) \
                                   .sort_values(by='order_id', ascending=False) \
                                   .reset_index()

total = count_status.order_id.sum()

figure = plt.figure(figsize=(20,5))
ax = sns.barplot(data = count_status, x= 'order_status', y= 'order_id')
ax.set_title("\n Quantidade de pedidos por status | Total de pedidos = {} \n".format(total))
ax.set(xlabel='Status', ylabel='Pedidos')
ax.tick_params(axis='x', rotation=45)
ax.plot();
```

```python
percentage_not_delivered = total_not_delivered / total

print("Percentual de pedidos não entregues: {:.2%}".format(percentage_not_delivered))
```

#### VAMOS ANALISAR OS PEDIDOS REALIZADOS AO LONGO DO TEMPO
<BR>

Para isso, criaremos algumas novas colunas derivadas da coluna **order_purchase_timestamp**

```python
# transformando a coluna order_purchase_timestamp de string para datetime
olist_orders_dataset['order_purchase_timestamp'] = olist_orders_dataset['order_purchase_timestamp'].astype('datetime64[ns]')

# criando novas colunas a partir da coluna order_purchase_timestamp
olist_orders_dataset['order_purchase_date'] = olist_orders_dataset.order_purchase_timestamp.dt.date.astype('datetime64[ns]')
olist_orders_dataset['order_purchase_datemonth'] = olist_orders_dataset.order_purchase_timestamp.dt.strftime('%Y-%m-01').astype('datetime64[ns]')
olist_orders_dataset['order_purchase_month'] = olist_orders_dataset.order_purchase_timestamp.dt.month.map({1:'Janeiro',2:'Fevereiro',3:'Março',4:'Abril',5:'Maio',6:'Junho',7:'Julho',8:'Agosto',9:'Setembro',10:'Outubro',11:'Novembro',12:'Dezembro'})
olist_orders_dataset['order_purchase_weekday'] = olist_orders_dataset.order_purchase_timestamp.dt.weekday.map({0:'Segunda',1:'Terça',2:'Quarta',3:'Quinta',4:'Sexta',5:'Sabado',6:'Domingo'})
olist_orders_dataset['order_purchase_hour'] = olist_orders_dataset.order_purchase_timestamp.dt.hour
olist_orders_dataset['order_purchase_year'] = olist_orders_dataset.order_purchase_timestamp.dt.year
olist_orders_dataset['order_purchase_quarter'] = olist_orders_dataset.order_purchase_timestamp.dt.quarter.map({1:'1º Trimestre',2:'2º Trimestre',3:'3º Trimestre',4:'4º Trimestre'})
```

```python
by = 'order_purchase_datemonth'

agg = (olist_orders_dataset.groupby([by]).agg({'order_id':'count'}).sort_values(by=by, ascending=True))

figure = plt.figure(figsize=(20,5))
ax = sns.lineplot(data=agg, x=by, y='order_id', linewidth=4)
ax.set(xlabel='Data da compra', ylabel='Pedidos')
ax.tick_params(axis='x', rotation=45)
ax.set_title("\n Evolução das vendas de Ecommerce no Brasil por mês \n")
ax.plot();
```

> Parece haver algo estranho nos dados referentes a 2016 e nos dados de setembro e outrobro de 2018.
> 
> Provavelmente isso é algum ruido da base, então vamos remover esses período e seguir com a analise.

```python
orders = olist_orders_dataset[(olist_orders_dataset.order_purchase_date >= '2017-01-01') & (olist_orders_dataset.order_purchase_date <= '2018-08-31')]
```

```python
by = 'order_purchase_datemonth'

agg = orders.groupby([by]).agg({'order_id':'count'}).sort_values(by=by, ascending=True)

_max = agg[agg['order_id']==agg['order_id'].max()].reset_index()

figure = plt.figure(figsize=(20,5))
ax = sns.lineplot(data=agg, x=by, y='order_id', linewidth=3)
ax.tick_params(axis='x', rotation=45)
ax.set(xlabel='Data da compra', ylabel='Pedidos')
ax.set_title("\n Evolução dos pedidos de vendas de Ecommerce no Brasil por mês (Jan/2017 - Ago/2018) \n")
plt.scatter(data=_max, x=by, y='order_id', c='r', linewidths=4)
ax.plot();
```

```python
by = 'order_purchase_date'

agg = orders.groupby([by]).agg({'order_id':'count'}).sort_values(by=by, ascending=True)

_max = agg[agg['order_id']==agg['order_id'].max()].reset_index()

figure = plt.figure(figsize=(20,5))
ax = sns.lineplot(data=agg, x=by, y='order_id', linewidth=3)
ax.tick_params(axis='x', rotation=45)
ax.set(xlabel='Data da compra', ylabel='Pedidos')
ax.set_title("\n Evolução dos pedidos vendas de Ecommerce no Brasil por dia (Jan/2017 - Ago/2018) \n")

plt.scatter(data=_max, x=by, y='order_id', c='r', linewidths=3);
```

> Podemos perceber que há um crecimento no numero de pedidos no Brasil nesse periodo de 20 meses.
> 
> É notável o pico de vendas com **mais de 1.000 pedidos** em 24 de novembro de 2017, quando houve a **Blackfriday**.
> 
> Vamos aplicar uma média móvel de 12 meses para suavizar a série e ver se conseguimos detectar algum padrão.

```python
by = 'order_purchase_date'

agg = orders.groupby([by]).agg({'order_id':'count'}).sort_values(by=by, ascending=True)
agg = agg.rolling(window=12).mean().reset_index()

figure = plt.figure(figsize=(20,5))
ax = sns.lineplot(data=agg, x=by, y='order_id', linewidth=3)
ax.tick_params(axis='x', rotation=45)
ax.set(xlabel='Data da compra', ylabel='Pedidos')
ax.set_title("\n Média móvel de 12 meses dos pedidos de vendas de Ecommerce no Brasil (Jan/2017 - Ago/2018) \n")

_max = agg[agg['order_id']==agg['order_id'].max()].reset_index()
plt.scatter(data=_max, x=by, y='order_id', c='r', linewidths=3)
ax.plot();
```

```python
by = 'order_purchase_month'

agg_day = (orders.groupby(['order_purchase_date', by])
               .agg({'order_id':'count'})
               .sort_values(by='order_purchase_date', ascending=True).reset_index())

agg = (agg_day.groupby([by])
               .agg({'order_id':'median'})
               .sort_values(by='order_id', ascending=False).reset_index())

figure = plt.figure(figsize=(10, 7))
ax = sns.barplot(data=agg, y=by, x='order_id')
ax.set(xlabel='Pedidos', ylabel='Mês da compra')
ax.tick_params(axis='x', rotation=0)
ax.set_title("\n Médianas de pedidos de Ecommerce no Brasil por mês do ano (Jan/2017 - Ago/2018) \n");
```

```python
by = 'order_purchase_month'

agg_day = (orders.groupby(['order_purchase_date', by])
               .agg({'order_id':'count'})
               .sort_values(by='order_purchase_date', ascending=True).reset_index())

agg = (agg_day.groupby([by])
               .agg({'order_id':'median'})
               .sort_values(by='order_id', ascending=False).reset_index())

figure = plt.figure(figsize=(10, 7))
ax = sns.barplot(data=agg, y=by, x='order_id')
ax.set(xlabel='Pedidos', ylabel='Mês da compra')
ax.tick_params(axis='x', rotation=0)
ax.set_title("\n Médianas de pedidos de Ecommerce no Brasil por mês do ano (Jan/2017 - Ago/2018) \n");
```

```python
by = 'order_purchase_weekday'

agg_day = orders.groupby(['order_purchase_date', by]) \
               .agg({'order_id':'count'}) \
               .sort_values(by='order_purchase_date', ascending=True).reset_index()

figure = plt.figure(figsize=(20, 7))
ax = sns.boxplot(data=agg_day, x=by, y='order_id', showfliers=False)
ax.set(xlabel='Dia da compra', ylabel='Pedidos')
ax.tick_params(axis='x', rotation=0)
ax.set_title("\n Boxplot dos pedidos de Ecommerce no Brasil por dia da semana (Jan/2017 - Ago/2018) \n")
ax.plot();
```

```python
by = 'order_purchase_hour'

agg_day = (orders.groupby(['order_purchase_date', by])
               .agg({'order_id':'count'})
               .sort_values(by='order_purchase_date', ascending=True).reset_index())

agg = (agg_day.groupby([by])
               .agg({'order_id':'median'})
               .sort_values(by=by, ascending=True).rename(columns={'order_id':'Qtd Pedidos'}).reset_index())

# graph
figure = plt.figure(figsize=(20,5))
ax = sns.barplot(data=agg, x=by, y='Qtd Pedidos')
ax.tick_params(axis='x', rotation=0)

# labels
for i in agg.index:
    y = round(agg.iloc[i].values[1],1)
    x = i-0.2
    ax.text(x=x, y=y-.5, s=y, va = 'center', c='w')

# title    
ax.set_title("\n Médianas de pedidos de Ecommerce no Brasil por hora do dia (Jan/2017 - Ago/2018) \n")
ax.plot();
```

### INSIGHTS
**Analisando os pedidos de ecommerce no Brasil no período de jan/2017 à ago/2018, podemos resumir que:**

* Existe uma tendencia de crescimento nas compras online no Brasil ao longo dos anos
* Os melhores meses de vendas são novembro (blackfriday) e dezembro (natal)

```python
products = olist_products_dataset.copy()
products.head(5)
```

A tabela de cadastro de produtos, nos informa as dimensões de cada produto, o que pode ser útil, porém a tabela já contêm muitos campos.

Buscando uma forma de reduzir o numero de colunas, podemos calcular o peso cúbico de cada produto. Seguiremos o padrão dos correios. 
> Mais informações sobre peso cúbico:
> * https://ajuda.melhorenvio.com.br/pt-BR/articles/4640361-como-e-calculado-o-peso-cubico-pelos-correios
> * https://blog.melhorenvio.com.br/a-importancia-da-cubagem-no-frete-para-e-commerce/

```python
products['product_weight_m3'] = ((products.product_length_cm * products.product_height_cm * products.product_width_cm) / 6000).round(3)
products.head()
```

```python
products[['product_weight_g', 'product_weight_m3']].corr(method='pearson').round(3)
```

Podemos ver que existe de fato uma correlação positiva forte entre o peso do produto e o peso cúbico.

Provavelmente iremos selecionar apenas um desses campos para manter no dataset devido a multicolinearidade, mas por enquanto seguiremos com os dois

```python
products.head()
```

Como podemos ver, nessa tabela a coluna **geolocation_zip_code_prefix** não é um ID unico, pois contem varias coordenadas diferentes.
Também iremos desconsiderar a informação de **geolocation_city**, pois além de os nomes não estarem padronizados, ao pegar a coordenada média por estado, não faz sentido ter a cidade.

Vamos pegar uma coordenada média para cada código, isso irá facilitar nossa analise.

```python
geolocation = (
    olist_geolocation_dataset.
    groupby(['geolocation_zip_code_prefix', 'geolocation_state']).
    agg({'geolocation_lat':'mean', 'geolocation_lng': 'mean'}).
    reset_index())

geolocation.head()
```

Nessa tabela temos a informação de nome da cidade e estado por CEP.

##### * Região do País

Vamos adicionar a informação de região do país ao qual cada estado pertence.

Para isso vamos extrair a informação do site: https://www.estadosecapitaisdobrasil.com/

```python
import requests
from bs4 import BeautifulSoup as bs

uri = requests.get('https://www.estadosecapitaisdobrasil.com/')

if uri.status_code == 200:
    try:
        soup = bs(uri.text, 'html.parser')

        table_class = "table tabela-estados-index tabela-estados table-striped text-center table-responsive"

        table = soup.find(class_=table_class)
        rows = table.find_all('tr')
        rows = rows[1:]

        regions = []
        for row in rows:
            geolocation_state  = row.contents[5].get_text().replace('<td>', '').replace('</td>', '')
            geolocation_region = row.contents[9].get_text().replace('\n','').replace(' ','')

            regions.append(tuple((geolocation_state, geolocation_region)))

        regions = pd.DataFrame(data = regions, columns=['geolocation_state', 'geolocation_region'])
    except Exception as e:
        print(e)
else:
    print("Falha ao conectar.")

regions.head()
```

```python
geolocation = geolocation.merge(regions)
geolocation.sample(5)
```

```python
def plot_pareto_by(df_, group_by, column):

    df = df_.groupby(group_by)[column].sum().reset_index()
    df = df.sort_values(by=column,ascending=False)

    df['cumpercentage'] = df[column].cumsum()/df[column].sum()*100
#     df['const'] = 1
#     df['cum_perc'] = df['const'].cumsum() / len(df) * 100
#     df = df.head(25)
    
    fig, ax = plt.subplots(figsize=(20,8))
    ax.bar(df[group_by], df[column], color="C0")
#     ax.axhline(20, color="orange", linestyle="dashed")
    
    ax2 = ax.twinx()
    ax2.plot(df[group_by], df["cumpercentage"], color="C1", marker="D", ms=6)
    ax2.yaxis.set_major_formatter(PercentFormatter())
    ax2.axhline(80, color="orange", linestyle="dashed")

    ax.tick_params(axis="y", colors="C0")
    ax2.tick_params(axis="y", colors="C1")

    for tick in ax.get_xticklabels():
        tick.set_rotation(90)
    plt.show()
```

```python
def abc_curve(cumpercentage):
    if cumpercentage < 81:
        ABC_curve = 'A'
    elif cumpercentage < 96:
        ABC_curve = 'B'
    else:
         ABC_curve = 'C'
    return ABC_curve

agg =items.groupby([by]).agg({'price':'sum'}).sort_values(by='price', ascending=False)
agg['percentage'] = agg['price'] / agg['price'].sum() * 100
agg['cumpercentage'] = agg['percentage'].cumsum()
agg = agg.sort_values(by='cumpercentage', ascending=True)
agg['const'] = 1
agg['acum_cat'] = agg['const'].cumsum() / len(agg) * 100

agg['ABC_curve'] = agg.apply(lambda agg: abc_curve(agg.cumpercentage), axis=1)
agg = agg.round()
```

##### * Distancia entre o remetente e o destinatário

Para calcular a distancia vamos precisar importar a base **olist_geolocation_dataset** pata obter as coordenadas exatas

Importaremos também a biblioteca **geopy** para conseguir obter a distancia entre as coordenadas

```python
from geopy.distance import great_circle

def AproxDistance(coord1, coord2):
    
    # 1.25 é um multiplicador para aproximar o valor considerando estradas e não uma linha reta
    roads_consider = 1.25
    try:
        distance = round((great_circle(coord1, coord2).km * roads_consider))
        return distance
    except ValueError:
        return np.nan        

def CoordsByCodePrefix(zip_code_prefix_1, zip_code_prefix_2):

    coord1 = (geolocation[geolocation['geolocation_zip_code_prefix']==zip_code_prefix_1].geolocation_lat.mean(),
              geolocation[geolocation['geolocation_zip_code_prefix']==zip_code_prefix_1].geolocation_lng.mean())

    coord2 = (geolocation[geolocation['geolocation_zip_code_prefix']==zip_code_prefix_2].geolocation_lat.mean(),
              geolocation[geolocation['geolocation_zip_code_prefix']==zip_code_prefix_2].geolocation_lng.mean())

    return coord1, coord2
```

```python
orders_merge['distance_seller_customer_km'] = (
    orders_merge.apply(lambda df:
        AproxDistance(
            CoordsByCodePrefix(df.seller_zip_code_prefix, df.customer_zip_code_prefix)[0],
            CoordsByCodePrefix(df.seller_zip_code_prefix, df.customer_zip_code_prefix)[1]), 
    axis=1)
)
```
