##### * Distancia entre o remetente e o destinatário

Para calcular a distancia vamos precisar importar a base **olist_geolocation_dataset** pata obter as coordenadas exatas

Importaremos também a biblioteca **geopy** para conseguir obter a distancia entre as coordenadas

```python
geolocation = pd.read_csv("g:/Meu Drive/Carreira/Estudos/Data Science/4 - Projetos/data-science/projects/brazilian-ecommerce/data/raw/olist_geolocation_dataset.csv")
from geopy.distance import great_circle

def AproxDistance(coord1, coord2):
    
    # 1.25 é um multiplicador para aproximar o valor considerando estradas e não uma linha reta
    roads_consider = 1.25
    try:
        distance = round((great_circle(coord1, coord2).km * roads_consider))
        return distance
    except ValueError:
        return np.nan        

def CoordsByCodePrefix(seller_zip_code_prefix, customer_zip_code_prefix):

    coord1 = (geolocation[geolocation['geolocation_zip_code_prefix']==seller_zip_code_prefix].geolocation_lat.mean(),
              geolocation[geolocation['geolocation_zip_code_prefix']==seller_zip_code_prefix].geolocation_lng.mean())

    coord2 = (geolocation[geolocation['geolocation_zip_code_prefix']==customer_zip_code_prefix].geolocation_lat.mean(),
              geolocation[geolocation['geolocation_zip_code_prefix']==customer_zip_code_prefix].geolocation_lng.mean())

    return coord1, coord2
```

```python
orders_delivered_items['distance_seller_customer_km'] = orders_delivered_items.apply(lambda df:
                                        AproxDistance(CoordsByCodePrefix(df.seller_zip_code_prefix, df.customer_zip_code_prefix)[0],
                                                      CoordsByCodePrefix(df.seller_zip_code_prefix, df.customer_zip_code_prefix)[1]), axis=1)
```

Outliers são registros que se diferenciam muito dos demais registros. São chamados de pontos fora da curva ou anomalias.
Os outliers podem influenciar negativamente nossa analise e portanto precisamos identificar e tratar esses pontos.

Existem diversos métodos para identificar e tratar outliers. 

Nessa analise vamos utilizar alguns gráficos e o método do **intervalo interquartil / amplitude interquartil** (IQR) para identifcá-los.

> IQR = Q3 - Q1
>
> limite inferior = Q1 - (1.5 * IQR)
>
> limite superior = Q3 + (1.5 * IQR)
>
>Quaisquer pontuações inferiores ao limite inferior ou superiores ao limite superior são considerados outliers.
>

Referencias: 
* https://medium.com/analytics-vidhya/outlier-treatment-9bbe87384d02
* https://www.datasciencesmachinelearning.com/2018/11/handling-outliers-in-python.html

Podemos perceber que existem produtos com valores muito acima da maioria. 

Vamos utilizar o método de IQR para identificar e analisar esses valores.

```python
def IQR_function(x):
    
    """
    A function to calculate the IQR, lower and upper bound \n
    Parameters
    ----------
    x : series_like \n
    Returns the IQR, lower_bound and upper_bound (float or series)
    """
    
    Q1 = x.quantile(.25)
    Q3 = x.quantile(.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)

    return IQR, lower_bound, upper_bound
```

```python
IQR_price, lower_bound_price, upper_bound_price = IQR_function(orders_delivered_items.price)

price_outliers = orders_delivered_items[(orders_delivered_items.price < lower_bound_price) | (orders_delivered_items.price > upper_bound_price)]
price_without_outliers = orders_delivered_items[(orders_delivered_items.price >= lower_bound_price) & (orders_delivered_items.price <= upper_bound_price)]

percentage_outliers = len(price_outliers) / len(orders_delivered_items)

print('Quantidade de outliers: {}'.format(len(price_outliers)))
print('Percentual de outliers: {:.2%}'.format(percentage_outliers))
```

Como podemos observar, existem categorias de produtos com preços mais elevados, portanto optar remover os outliers dentro de cada categoria de produto e comparar com a remoção dos outliers sem considerar as categorias

```python
category_name_agg = orders_delivered_items.groupby('product_category_name')['price']
```

```python
IQR, lower_bound, upper_bound = IQR_function(category_name_agg)

orders_delivered_items_filtered = pd.DataFrame()
price_category_outliers = pd.DataFrame()

for category_name in category_name_agg.groups.keys():
    
    is_category = orders_delivered_items['product_category_name'] == category_name    
    is_bounded = (orders_delivered_items.price >= lower_bound[category_name]) & (orders_delivered_items.price <= upper_bound[category_name])
    is_outlier = (orders_delivered_items.price < lower_bound[category_name]) | (orders_delivered_items.price > upper_bound[category_name])
    
    selection = is_category & is_bounded  
    orders_delivered_items_selection = orders_delivered_items[selection]
    orders_delivered_items_filtered = pd.concat([orders_delivered_items_filtered, orders_delivered_items_selection])
    
    selection_outliers = is_category & is_outlier  
    price_category_selection_outliers = orders_delivered_items[selection_outliers]
    price_category_outliers = pd.concat([price_category_outliers, price_category_selection_outliers])    
```

```python
percentage_outliers = len(price_category_outliers) / len(orders_delivered_items)

print('Quantidade de outliers: {}'.format(len(price_category_outliers)))
print('Percentual de outliers: {:.2%}'.format(percentage_outliers))
```
